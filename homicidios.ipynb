```python
import pandas as pd
```

```python
Homicidios = pd.ExcelFile('C:\\Users\\Dr-Pc\\OneDrive\\Escritorio\\PROYECTO INDIVIDUAL N°2 Siniestros Viales\\homicidios.xlsx')
Homicidios_1 = pd.read_excel(Homicidios, 'HECHOS')
Homicidios_2 = pd.read_excel(Homicidios,'VICTIMAS')
```

```python
Homicidios_1.rename(columns={'ID': 'ID_hecho'}, inplace=True)
```

```python
Update_Homicidios = pd.merge(Homicidios_1, Homicidios_2, on='ID_hecho')
```

```python
Update_Homicidios.head()
```

```python
Update_Homicidios.rename(columns={'FECHA_x': 'FECHA'}, inplace=True)
Update_Homicidios.rename(columns={'AAAA_x': 'AAAA'}, inplace=True)
Update_Homicidios.rename(columns={'MM_x': 'MM'}, inplace=True)
Update_Homicidios.rename(columns={'DD_x': 'DD'}, inplace=True)
Update_Homicidios.rename(columns={'VICTIMA_y': 'VICTIMA'}, inplace=True)
Update_Homicidios.rename(columns={'FECHA_FALLECIMIENTO_x': 'FECHA_FALLECIMIENTO'}, inplace=True)
```

```python
Update_Homicidios.drop('Cruce' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('Calle' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('Dirección Normalizada' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('FECHA_y' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('AAAA_y' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('MM_y' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('DD_y' , axis = 1, inplace = True)
```

```python
Update_Homicidios.drop('VICTIMA_x' , axis = 1, inplace = True)
```

```python
Update_Homicidios['LUGAR_DEL_HECHO'] =Update_Homicidios['LUGAR_DEL_HECHO'].str.replace(',', '')
```

```python
Update_Homicidios.rename(columns={'HORA': 'HORA_DEL_SINIESTRO'}, inplace=True)
```

```python
Update_Homicidios['FECHA_FALLECIMIENTO'] = Update_Homicidios['FECHA']
```

```python
def reorganizar_direcc(address):
    words = address.split()
    if "Av." in words:
        words.remove("Av.")
        words = ["Av."] + words
    return " ".join(words)

Update_Homicidios['LUGAR_DEL_HECHO'] = Update_Homicidios['LUGAR_DEL_HECHO'].apply(reorganizar_direcc)
```

```python
# Convertir la columna 'FECHA' a objetos datetime
Update_Homicidios['FECHA'] = pd.to_datetime(Update_Homicidios['FECHA'], errors='coerce')

# Formatear la columna 'FECHA' para que muestre solo la fecha en 'YYYY-MM-DD'
Update_Homicidios['FECHA'] = Update_Homicidios['FECHA'].dt.strftime('%Y-%m-%d')
```

```python
# Convertir la columna 'FECHA' a objetos datetime
Update_Homicidios['FECHA_FALLECIMIENTO'] = pd.to_datetime(Update_Homicidios['FECHA_FALLECIMIENTO'], errors='coerce')

# Formatear la columna 'FECHA' para que muestre solo la fecha en 'YYYY-MM-DD'
Update_Homicidios['FECHA_FALLECIMIENTO'] = Update_Homicidios['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m-%d')
```

```python
Update_Homicidios.head()
```

```python
conteo_sd = Update_Homicidios.apply(lambda x: x.value_counts().get('Sd', 0))
#1900-01-14T07:12:00.000
formato_incorrecto = pd.to_datetime(Update_Homicidios['HORA_DEL_SINIESTRO'], errors='coerce').isnull()

#Convertir el formato ISO 8601 a HH:mm:ss
Update_Homicidios.loc[~formato_incorrecto, 'HORA_DEL_SINIESTRO'] = pd.to_datetime(Update_Homicidios.loc[~formato_incorrecto, 'HORA_DEL_SINIESTRO']).dt.strftime('%H:%M:%S')
```

```python
Update_Homicidios['pos x'] = Update_Homicidios['pos x'].replace('.', 0.0)
Update_Homicidios['pos y'] = Update_Homicidios['pos y'].replace('.', 0.0)

# Convertir a tipo de dato numérico (opcional, dependiendo de tu necesidad)
Update_Homicidios['pos x'] = pd.to_numeric(Update_Homicidios['pos x'], errors='coerce').astype(float)
Update_Homicidios['pos y'] = pd.to_numeric(Update_Homicidios['pos y'], errors='coerce').astype(float)
```
