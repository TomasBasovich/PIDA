# Utilizamos la extension data wrangler para ordenar el dataset scrappeado

```python

def clean_data(df):
    # Drop column: 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]​.7'
    df = df.drop(columns=['Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.7'])
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b' to 'puesto (2022)'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b': 'puesto (2022)'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.1' to 'Comuna'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.1': 'Comuna'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.2' to 'CENSO 1991'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.2': 'CENSO 1991'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.3' to 'CENSO 2001'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.3': 'CENSO 2001'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.4' to 'CENSO 2010'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.4': 'CENSO 2010'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.5' to 'CENSO 2022'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.5': 'CENSO 2022'})
    # Rename column 'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.6' to 'CRECIMIENTO 2010-2022'
    df = df.rename(columns={'Evolución demográfica e histórica de los habitantes de las Comunas de la Ciudad Autónoma de Buenos Aires (CABA)[1]\u200b.6': 'CRECIMIENTO 2010-2022'})
    # Drop column: 'CENSO 2001'
    df = df.drop(columns=['CENSO 2001'])
    # Drop column: 'CENSO 1991'
    df = df.drop(columns=['CENSO 1991'])
    return df

df = clean_data(df.copy())
df.head()
```

# Calculamos la tasa de crecimiento anual para poder tener datos de 2021 sobre el indice de crecimiento y poblacion total segun el censo realizado en 2022

```python
# Eliminamos la primer fila que contiene el titulo de las columnas
df = df.drop(df.index[0])

# Convertimos la columna 'CENSO 2022' a float
df['CENSO 2022'] = df['CENSO 2022'].str.replace('\xa0', '').astype(float)

# Poblacion total de 2022
censo_2022 = 3120612

# Tasa de crecimiento total entre 2010 y 2022
tasa_crecimiento_total = 7.97 / 100

# Calcular la tasa de crecimiento anual
tasa_crecimiento_anual = (1 + tasa_crecimiento_total) ** (1 / 12) - 1

# Crear una nueva columna con los valores estimados para 2021
df['CENSO 2021'] = df['CENSO 2022'] / (1 + tasa_crecimiento_anual)
```

# Observamos que ya tenemos el censo y poblacion de comunas de 2021, pero necesitamos los demas años, para ello vamos a realizar un calculo basado en la columna "CRECIMIENTO 2010-2022" el cual nos indica el indice en porcentajes de crecimiento que será el factor clave para el desarrollo exitoso de nuestro codigo

```python


# Datos de población de 2010 y tasas de crecimiento para nuestro calculo 
data = {
    'Comuna': ['Comuna 13', 'Comuna 14', 'Comuna 12', 'Comuna 4', 'Comuna 1', 'Comuna 7', 'Comuna 8', 'Comuna 6', 'Comuna 11', 'Comuna 15', 'Comuna 3', 'Comuna 5', 'Comuna 10', 'Comuna 9', 'Comuna 2'],
    'Poblacion 2010': [231331, 225970, 200116, 218245, 205886, 220591, 187237, 176076, 189832, 182574, 187537, 179005, 166022, 161797, 157932],
    'Crecimiento 2010-2022': [14.64, 10.19, 18.07, 5.81, 8.44, -1.70, 9.40, 15.73, 7.19, 8.39, 4.22, 8.29, 3.47, 4.78, 0.27]
}

df = pd.DataFrame(data)

# Convertir tasas de crecimiento en factores de crecimiento
df['Factor de crecimiento'] = 1 + df['Crecimiento 2010-2022'] / 100

# Calcular población para cada año
for year in range(2016, 2022):
    n = year - 2010
    df[f'Poblacion {year}'] = df['Poblacion 2010'] * df['Factor de crecimiento']**(n/12)

df
```
